## Random Forest

**Definition:** Random Forest is an ensemble learning method that constructs a multitude of decision trees at training time. For classification tasks, it outputs the class that is the mode of the classes (majority vote) of the individual trees. For regression tasks, it outputs the mean prediction of the individual trees.

**Why and Where We Use It:**
*   **High Accuracy:** It generally provides high accuracy and works well on a large range of datasets, especially those with many features and complex interactions.
*   **Robustness to Overfitting:** By averaging the predictions of multiple trees, it reduces the risk of overfitting compared to a single decision tree.
*   **Feature Importance:** It can implicitly measure the importance of features, which helps in feature selection.
*   **Handles Missing Values:** It can handle missing values and maintains accuracy for a large proportion of missing data.
*   **Non-linear Relationships:** It can model non-linear relationships between features and target variables.

**Common Use Cases:**
*   **Classification:** Image classification, medical diagnosis, fraud detection, customer churn prediction.
*   **Regression:** Stock price prediction, demand forecasting, drug response prediction.

**Advantages:**
*   Reduces overfitting compared to single decision trees.
*   Works well with both categorical and numerical data.
*   Provides good estimates of feature importance.
*   Can handle large datasets with high dimensionality.
*   Maintains accuracy even with a large proportion of missing data.

**Limitations:**
*   **Interpretability:** Less interpretable than a single decision tree due to the ensemble nature.
*   **Computational Cost:** Can be computationally expensive and slower to train than other algorithms, especially with a large number of trees and features.
*   **Resource Intensive:** Requires more memory as it stores multiple decision trees.
*   **Bias:** Can be biased towards features with more categories when dealing with categorical variables.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np

x=np.array([[1],[2],[3],[4],[5],[6]])
y=np.array([4.4,8.6,9.4,5.8,4.4,5.6])

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
model=RandomForestRegressor(n_estimators=10)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(r2_score(y_test,y_pred))
```
