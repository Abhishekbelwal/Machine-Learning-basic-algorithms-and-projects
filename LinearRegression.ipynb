```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

iris = load_iris()
X_iris = iris.data
y_iris = iris.target

model_iris = LinearRegression()
model_iris.fit(X_iris, y_iris)

y_pred_iris = model_iris.predict(X_iris)

r2_iris = r2_score(y_iris, y_pred_iris)

print(f"R2 Score for Iris dataset: {r2_iris}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris_df = pd.DataFrame(data=X_iris, columns=iris.feature_names)
iris_df['species'] = y_iris

iris_df['species'] = iris_df['species'].map(lambda x: iris.target_names[x])

fig = sns.pairplot(iris_df, hue='species', palette='viridis')
plt.suptitle('Pair Plot of Iris Dataset Features by Species', y=1.02) # Add a title
plt.show()
```
