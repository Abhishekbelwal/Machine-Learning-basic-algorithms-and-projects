```r
install.packages("rpart")
library (rpart)
```

```r
data<-data.frame(
  hours=c(1,2,3,4,5,6,7,8),
  attendence=c(40,45,50,55,60,65,70,75),
  pass=as.factor(c("no","yes","yes","no","no","yes","no","yes"))
)
```

```r
model<-rpart(pass~hours+attendence,data=data,method="class", control=rpart.control(minsplit=2, cp=0))
pred<-predict(model,data,type="class")
print(pred)
```

```r
table(Predicted=pred,Actual=data$pass)
```
