Polynomial Regression using R


```r
data<-data.frame(level=1:10,Salary=c(45000,50000,55000,60000,65000,87000,70000,10000,450000,740000))
model<-lm(Salary~poly(level,2),data=data)
predicted<-predict(model,data)
print(predicted)
plot(data$level,data$Salary,main="Polynomial Regression",pch=16,xlab="Level",ylab="salary",col="blue")
lines(data$level,predicted,col="red",lwd=2)
```
