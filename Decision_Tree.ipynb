### What is a Decision Tree?

A **Decision Tree** is a supervised machine learning algorithm that can be used for both classification and regression tasks. It builds a model in the form of a tree structure, where each internal node represents a 'test' on an attribute (e.g., is the temperature > 20°C?), each branch represents the outcome of the test, and each leaf node represents a class label (in classification) or a numerical value (in regression).

### Why use Decision Trees?

1.  **Interpretability:** Decision trees are very easy to understand and interpret. The tree structure mimics human decision-making, making it simple to visualize and explain the logic behind a prediction.
2.  **No Data Preprocessing Needed:** They generally don't require feature scaling or normalization, unlike many other algorithms (like SVM or KNN).
3.  **Handles Both Numerical and Categorical Data:** They can work with both types of features without special conversions.
4.  **Handles Non-linear Relationships:** Decision trees can capture complex non-linear relationships between features and the target variable.

### Where are Decision Trees used?

Decision trees are widely used in various fields:

*   **Medical Diagnosis:** To predict the likelihood of a disease based on symptoms and patient history.
*   **Customer Relationship Management (CRM):** To predict customer churn, identify potential customers, or personalize marketing campaigns.
*   **Financial Analysis:** For credit scoring, fraud detection, and predicting stock price movements.
*   **Manufacturing:** To identify defects in products or optimize production processes.
*   **Image Recognition:** Although deep learning is more common, decision trees can be used for simpler image classification tasks.

In essence, decision trees are excellent for situations where you need a clear, interpretable model that can handle various data types without extensive preprocessing.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np
x=np.array([[1],[2],[3],[4],[5],[6]])
y=np.array([1.4,1.6,1.4,1.8,4.4,5.6])

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

model=DecisionTreeRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(r2_score(y_test,y_pred))
```
