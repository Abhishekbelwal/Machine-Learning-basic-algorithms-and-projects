```r
library(e1071)
```

```r
data<-data.frame(Exp=c(1,2,3,4,5,6,7),Sal=c(100,20,300,490,500,760,700))
svr_model<-svm(Sal~Exp,data=data,type = 'eps-regression',kernel='radial')
new_data<-data.frame(Exp=c(2.5,5.5))
pred_sal<-predict(svr_model,new_data)
print(pred_sal)
plot(data$Exp, data$Sal, col='blue', main="svr")
lines(data$Exp,predict(svr_model,data),col='red',lwd=2,)
```
