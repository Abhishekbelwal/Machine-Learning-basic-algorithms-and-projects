## Polynomial Regression: Definition, Equation, and Examples

**Definition:** Polynomial regression is a form of regression analysis used when the relationship between the independent variable (`x`) and the dependent variable (`y`) is non-linear. It models this relationship by fitting an _n_th-degree polynomial function to the data. While the relationship between `x` and `y` is curvilinear, it's still considered a linear model in terms of its coefficients, as it transforms `x` into polynomial features (e.g., $x, x^2, x^3$).

**Equation:**
The general mathematical form for a polynomial regression model of degree `n` is:

$y = \beta_0 + \beta_1x + \beta_2x^2 + \dots + \beta_nx^n + \epsilon$

Where:
*   `y` is the dependent variable (what you are trying to predict).
*   `x` is the independent variable (the predictor).
*   $\beta_0, \beta_1, \dots, \beta_n$ are the coefficients of the polynomial.
*   $\epsilon$ represents the error term.

**Examples of Use:**

1.  **Growth Modeling:** Predicting the growth of biological populations over time, where initial growth might be slow, then accelerate, and finally plateau.
2.  **Epidemiology:** Modeling the spread of a disease, which often follows a non-linear S-shaped curve.
3.  **Engineering:** Analyzing the performance of a material under varying stress, where the relationship isn't simply linear.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures

x=np.array([[1],[2],[3],[4],[5]])
y=np.array([1.2,4.5,6,7.8,9.2])
poly=PolynomialFeatures(degree=3)
x_poly=poly.fit_transform(x)

model=LinearRegression()
model.fit(x_poly,y)
pred=model.predict(x_poly)
print(r2_score(y,pred))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Original Data', color='blue')
plt.plot(x, pred, label='Polynomial Regression Curve', color='red', linestyle='--')
plt.title('Polynomial Regression Fit')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.legend()
plt.grid(True)
plt.show()
```
